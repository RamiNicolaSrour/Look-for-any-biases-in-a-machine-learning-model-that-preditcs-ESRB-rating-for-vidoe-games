Video games are rated and classified before release into the age groups that they are suitable, as Some games might include elements that aren't appropriate for children or people under a specific age.

In this notebook, i will build several machine learning models that classify games to the appropriate age group, games are from both xbox_one and PS4, and then see if the machine learning models have any discrimination against a certain gaming platform, and see if i can optimize any of the models

```python
# Imprt needed libraries to evaluate mahcine learning models, dealing with tables and datasets and preprocessing data
import pandas as pd
import sklearn.metrics
import sklearn.preprocessing
import sklearn.model_selection
```

```python
# setting python to show all columns, and then upload one of the datasets
pd.set_option('display.max_columns', None)
df1 = pd.read_csv("test_esrb.csv")
df1.head(5)
```

```python
# upload the next
df2 = pd.read_csv("Video_games_esrb_rating.csv")
df2.head(5)
```

```python
# datasets have same columns but different games so i will merge both datasets
# ignore_index is true to make new indices for the data
df = pd.concat([df1, df2], ignore_index=True)
df
```

# Data preprocessing

```python
# see if any nan values exist, if any exsit , it's better to delete them
Non_Values = df.isnull().sum()
Non_Values
```

```python
# checking the types of the data in the column, to see the columns that need encoding
df.dtypes
```

```python
# see how many unique values in the non numerical columns
# M is for Mature, T is for Teens, E for everyone, ET is to E+10 which means for ages 10 and up
uniquerate = df['esrb_rating'].unique().tolist()
uniquerate
```

```python
# turning the values in the esrb_rating to numbers
df['esrb_rating'] = df['esrb_rating'].replace({'E' : 0, 'ET' : 1, 'T' : 2, 'M' : 3})
df.head(3)
```

```python
# splitting the dataset to train and test
df_train, df_test = sklearn.model_selection.train_test_split(df)
print("number of columns and rows of the splitted data:", df_train.shape, df_test.shape)
```

```python
# check for any duplicate rows in the train set to delete them
duplicate_data = df_train.duplicated().sum()
duplicate_data
```

```python
#i have few duplicate rows in the train set, and i will remove them
df_train = df_train.drop_duplicates()
df_train.head(3)
```

```python
# dropping the title column since it doesn't effect the result
# i am dropping it now because, if i did it before i might have got more duplicate values
# duplicate values resulting from other columns be exact with only title being different
df_test = df_test.drop(columns = 'title', axis=1)
df_train = df_train.drop(columns = 'title', axis=1)
```

```python
# i will split the train and test datasets based on the console
# 0 is for Playstation4 games, 1 if for games that also works on Xbox one
xbox_df_train = df_train[df_train['console'] == 1]
PS4_df_train = df_train[df_train['console'] == 0]
xbox_df_test = df_test[df_test['console'] == 1]
PS4_df_test = df_test[df_test['console'] == 0]
```

```python
# splitting target label from the xbox train and test sets

xbox_x_train = xbox_df_train.drop(["esrb_rating"], axis = 1)
xbox_y_train = xbox_df_train["esrb_rating"]
xbox_x_test = xbox_df_test.drop(["esrb_rating"], axis = 1)
xbox_y_test = xbox_df_test["esrb_rating"]

# splitting target label from the ps4 train and test sets
PS4_x_train = PS4_df_train.drop(["esrb_rating"], axis = 1)
PS4_y_train = PS4_df_train["esrb_rating"]
PS4_x_test = PS4_df_test.drop(["esrb_rating"], axis = 1)
PS4_y_test = PS4_df_test["esrb_rating"]
```

# trying logistic regression

i will first try with logistic regression because it's one of the simple methods

```python
# building logistic regression
from sklearn.linear_model import LogisticRegression
LOG_Model = sklearn.linear_model.LogisticRegression()
```

#Disparate Mistreatment

Disparate Mistreatment happen when a model make more false predictions (or less true positives) between 2 different classes

i will check for accuracy, precision, recall and f1

```python
LOG_Model.fit(PS4_x_train, PS4_y_train)
ps4_y_predicted = LOG_Model.predict(PS4_x_test)
accuracy1 = sklearn.metrics.accuracy_score(PS4_y_test, ps4_y_predicted)
precision1, recall1, f11, support1= sklearn.metrics.precision_recall_fscore_support(PS4_y_test, ps4_y_predicted)
print(accuracy1, precision1, recall1, f11)
```

```python
LOG_Model.fit(xbox_x_train, xbox_y_train)
xbox_y_predicted = LOG_Model.predict(xbox_x_test)
accuracy2 = sklearn.metrics.accuracy_score(xbox_y_test, xbox_y_predicted)
precision2, recall2, f12, support2= sklearn.metrics.precision_recall_fscore_support(xbox_y_test, xbox_y_predicted)
print(accuracy2, precision2, recall2, f12)
```

the model seems to not discriminate between the 2 consoles, there is only some few differnces between the recall and f1 values

#Disparate impact
Disaprate impact is when the positive values between 2 groups of consoles are differnet.
I will see the Confusion matrix of both groups, and see if the positive values are very different.

```python
cm_PS4 = sklearn.metrics.confusion_matrix(PS4_y_test, ps4_y_predicted)
cm_xbox = sklearn.metrics.confusion_matrix(xbox_y_test, xbox_y_predicted)
print (cm_PS4,cm_xbox)
```

here also there is not much different, 2nd and 3rd clases are better fir ps4 and 4th class is better for xbox-one

#Disparate treatment
this happens when the prediction could change solely when the category column change (in our case the console)
for this i will switch the values of the consoles in the column and see how the results will be different.

```python
#make a dataset where the values represnting ps4 and xbox are switched
df_switched = df.copy()
df_switched['console'] = df_switched['console'].replace({1: 0, 0: 1})

# do the same splitting dataset and preprocessing steps i did before

df_train2, df_test2 = sklearn.model_selection.train_test_split(df_switched)

df_train2 = df_train2.drop_duplicates()
df_test2 = df_test2.drop(columns = 'title', axis=1)
df_train2 = df_train2.drop(columns = 'title', axis=1)

xbox_df_train2 = df_train2[df_train2['console'] == 1]
PS4_df_train2 = df_train2[df_train2['console'] == 0]
xbox_df_test2 = df_test2[df_test2['console'] == 1]
PS4_df_test2 = df_test2[df_test2['console'] == 0]


PS4_x_train2 = PS4_df_train2.drop(["esrb_rating"], axis = 1)
PS4_y_train2 = PS4_df_train2["esrb_rating"]
PS4_x_test2 = PS4_df_test2.drop(["esrb_rating"], axis = 1)
PS4_y_test2 = PS4_df_test2["esrb_rating"]
xbox_x_train2 = xbox_df_train2.drop(["esrb_rating"], axis = 1)
xbox_y_train2 = xbox_df_train2["esrb_rating"]
xbox_x_test2 = xbox_df_test2.drop(["esrb_rating"], axis = 1)
xbox_y_test2 = xbox_df_test2["esrb_rating"]
```

```python
# try to find disprate impact and mistreatment

LOG_Model.fit(PS4_x_train2, PS4_y_train2)
PS4_y_predicted2 = LOG_Model.predict(PS4_x_test2)
accuracy = sklearn.metrics.accuracy_score(PS4_y_test2, PS4_y_predicted2)
precision, recall, f1, support= sklearn.metrics.precision_recall_fscore_support(PS4_y_test2, PS4_y_predicted2)
print(accuracy, precision, recall, f1)
```

```python
LOG_Model.fit(xbox_x_train2, xbox_y_train2)
xbox_y_predicted2 = LOG_Model.predict(xbox_x_test2)
accuracy2 = sklearn.metrics.accuracy_score(xbox_y_test2, xbox_y_predicted2)
precision2, recall2, f12, support2= sklearn.metrics.precision_recall_fscore_support(xbox_y_test2, xbox_y_predicted2)
print(accuracy2, precision2, recall2, f12)
```

after swtiching the classes, we can also see that there is not much difference in accuracy.

```python
xbox_confusion2 = sklearn.metrics.confusion_matrix(xbox_y_test2, xbox_y_predicted2)
ps4_confusion2 = sklearn.metrics.confusion_matrix(PS4_y_test2, PS4_y_predicted2)
print(ps4_confusion2,xbox_confusion2 )
```

there isn't much differnce in predicitons between now and before seitihving the console values.

overall the model is good.
